## train 원핫인코딩

```python
df_train = pd.read_csv('data/train_44_frozen_.csv')
df_train.head()
```

```python
# '휴일종류' 열을 원핫인코딩하기
one_hot_encoded = pd.get_dummies(df_train['휴일종류'])
df_total = pd.concat([df_train, one_hot_encoded], axis=1)
df_total = df_total.drop(['휴일종류'], axis=1)
print(df_total.shape)
df_total.head()
```

```python
df_total.to_csv('train_onehot.csv', index=False)
```

## label 인코딩

```python
df_train = pd.read_csv('data/train_44_frozen_.csv')
df_train.head()
```

```python
df_test_1_10 = pd.read_csv('data/test_1_10_.csv')
df_test_1_10.head()
```

```python
df_test_1_15 = pd.read_csv('data/test_1_15_.csv')
df_test_1_15.head()
```

```python
df_test_2_10 = pd.read_csv('data/test_2_10_.csv')
df_test_2_10.head()
```

```python
df_test_2_15 = pd.read_csv('data/test_2_15_.csv')
df_test_2_15.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# train 데이터셋 레이블인코딩
label_encoder = LabelEncoder()
df_train['휴일종류_레이블'] = label_encoder.fit_transform(df_train['휴일종류'])

# df_test_1_10 레이블인코딩
df_test_1_10['휴일종류_레이블'] = label_encoder.transform(df_test_1_10['휴일종류'])

# df_test_1_15 레이블인코딩
df_test_1_15['휴일종류_레이블'] = label_encoder.transform(df_test_1_15['휴일종류'])

# df_test_2_10 레이블인코딩
df_test_2_10['휴일종류_레이블'] = label_encoder.transform(df_test_2_10['휴일종류'])

# df_test_2_15 레이블인코딩
df_test_2_15['휴일종류_레이블'] = label_encoder.transform(df_test_2_15['휴일종류'])
```

```python
df_train = df_train.drop(['휴일종류'], axis=1)

df_test_1_10 = df_test_1_10.drop(['휴일종류'], axis=1)
df_test_1_15 = df_test_1_15.drop(['휴일종류'], axis=1)
df_test_2_10 = df_test_2_10.drop(['휴일종류'], axis=1)
df_test_2_15 = df_test_2_15.drop(['휴일종류'], axis=1)
```

```python
df_train.to_csv('train_label.csv', index=False)

df_test_1_10.to_csv('test_1_10_label.csv', index=False)
df_test_1_15.to_csv('test_1_15_label.csv', index=False)
df_test_2_10.to_csv('test_2_10_label.csv', index=False)
df_test_2_15.to_csv('test_2_15_label.csv', index=False)
```
